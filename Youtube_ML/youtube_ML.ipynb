```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Sample initialization with adjusted parameters
vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform the data
tag_vectors = vectorizer.fit_transform(df['tags'])  # Assuming 'tags' column contains text data
```

```python
tag_normalization = tag_vectors.toarray()

tag_normalization
```

```python
# Choose the number of clusters (k) based on your data and requirements
k = 16 # Example number of clusters
kmeans = KMeans(n_clusters=k,random_state=10,n_init=10)
kmeans.fit(tag_normalization)

# Add the cluster labels to the DataFrame
df['cluster'] = kmeans.labels_
```

```python
df['cluster'].value_counts()
```

```python
df.head()
```

```python
score = silhouette_score(tag_normalization, kmeans.labels_)

score
```

```python
def recommend_videos(user_input, df, tag_vectors, vectorizer,top_n=500):
    # Transform the user input into a vector
    user_vector = vectorizer.transform([user_input])
    # Calculate cosine similarity between the user vector and video vectors
    similarities = cosine_similarity(user_vector, tag_vectors).flatten()

    # Get the indices of the top N most similar videos
    top_indices = similarities.argsort()[top_n:][::-1]
  


    # Retrieve the video IDs, channel titles, and titles of the top recommendations
    recommendations = df.iloc[top_indices][['video_id', 'channelTitle', 'title']]
    unique_recommendations = recommendations.drop_duplicates(subset='channelTitle')

    return unique_recommendations
```

```python
# Example user input
user_input = input('Enter your video name: ')

# Get recommendations
unique_recommendations = recommend_videos(user_input, df, tag_vectors, vectorizer)

# Print the recommendation
unique_recommendations
```
