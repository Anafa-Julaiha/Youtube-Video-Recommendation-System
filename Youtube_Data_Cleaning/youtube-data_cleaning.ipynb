```python
import pandas as pd
```

```python
#read json file
data1 = pd.read_json(r"C:\Users\Dell\Desktop\VS code\Youtube_project\data.json")
data2 = pd.read_json(r"C:\Users\Dell\Desktop\VS code\Youtube_project\data2.json")
```

```python
data1 = pd.read_csv(r"C:\Users\Dell\Desktop\VS code\Youtube_project\youtubedata1.csv")
data2 = pd.read_csv(r"C:\Users\Dell\Desktop\VS code\Youtube_project\youtubedata2.csv")
```

```python
#convert dataframe

d1 = pd.DataFrame(data1)
d2 = pd.DataFrame(data2)
```

```python
#Concat the 2 dataframes

df = pd.concat([d1,d2])
df
```

```python
#drop the columns [caption,definition,comment count, favourite count]
df.drop(['caption','definition','commentCount','favouriteCount','Unnamed: 0'],axis=1,inplace=True)
```

```python
# missing values
df[['tags','description','viewCount','likeCount']] = df[['tags','description','viewCount','likeCount']].fillna('')
```

```python
#convert string to numeric
df['viewCount'] = pd.to_numeric(df['viewCount'])
df['likeCount'] = pd.to_numeric(df['likeCount'])
```

```python
#convert tags uppercase to lowercase
df['tags'] = df['tags'].str.lower()
```
